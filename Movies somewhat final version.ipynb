```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\alexf\Downloads\movies.csv')
```

```python
# Now let's take a look at the data

df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
df['gross'] = df['gross'].astype('int64') 

df['budget'] = df['budget'].astype('int64') 
```

```python
df
```

```python
# We want to group by year as well for some of this as well as visualizations later so let's get the year from the released date
# We need to convert to string first before we do this

df['year'] = df['released'].astype(str).str[:4]

df
```

```python
# If there are any duplicates let's drop them

df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Now we have out data how we want it and we are ready to get started
```

```python
# We are going to be looking to see what variables effect the gross earnings of a movie

# We will be doing this mostly through correlation matrixs and plotting the data
```

```python
# We aren't going to look at every field and see how it looks against gross, but I want to focus on 3 that I think will
# be a big impact

# 1. I want to look at the Budget - I think the bigger the budget the more the movie will earn - smaller budget = smaller gross
# 2. 
# 3. Look at the Company - I think some companies have a high correlation to how much
```

```python
# Create a scatterplot of budget vs gross earnings

plt.scatter(x=df['budget'], y=df['gross'], alpha=0.9) # Alpha just changes the hue of the red dots
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Let's look at a plot of budget vs gross to see if the budget had any 

sns.regplot(x="budget", y="gross", data=df)
```

```python
correlation_matrix = df.corr() #methods are pearson, kendall, and spearman

sns.heatmap(correlation_matrix, annot = True) #annot shows the value of the correlation

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# There are different methods that you can use and each gives their own output
# I'll try to find a link in the description if you want to look into that 

correlation_matrix = df.corr(method='spearman') #methods are pearson, kendall, and spearman

sns.heatmap(correlation_matrix, annot = True) #annot shows the value of the correlation

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()

```

```python
# From what we have seen it looks like there is a fairly high correlation between how much a company 
# spends on the movie and how much it actually earns
# If it spends nothing it most likely didn't earn a lot
# if it spends a lot it most likely earned a lot 
```
